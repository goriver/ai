```python
import pandas as pd
from fbprophet import Prophet

df = pd.read_csv('../dataset/example_wp_log.csv')
```

```python
'''
#pip install --upgrade pip
#pip install pystan
#pip install fbprophet --user
'''
```

```python
df.describe()
```

```python
df.tail()
```

```python
m = Prophet()
m.fit(df)
```

```python
future = m.make_future_dataframe(periods=365)
```

```python
future.tail()
```

```python
forecast = m.predict(future)
```

```python
forecast[['ds','yhat','yhat_lower','yhat_upper']].tail() # yhat : predict
```

```python
fig1 = m.plot(forecast)
```

```python
fig2 = m.plot_components(forecast)
```
